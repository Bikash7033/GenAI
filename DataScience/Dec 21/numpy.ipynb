```python
import numpy as np
import time
#python list
size = 1_000_000
list1 = list(range(size))  # list1 = [0,1,2,3,4,.....999999]
list2 = list(range(size))   # list2 = [0,1,2,3,4,.....999999]
result = []  # we are creating a empty list where we will save the values
start = time.time()  
for x,y in  zip(list1,list2):  # zip(list1, list2) -=> [(0,0),(1,1)....]
    result.append(x+y)
    #print(value)
    #print(result)
end = time.time()
print("python list addition time : ", end - start)

#numpy array
arr1 = np.array(list1)
arr2 = np.array(list2)
start = time.time()
result = arr1 + arr2 
print(result)
end = time.time()
print("python numpy addition time : ", end - start)





```
