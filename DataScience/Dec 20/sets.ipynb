A set in python is an ordered, mutable and unique collection of elements . It does allow duplicate values.


```python
my_list = [1, 2, 3, 2, 4, 1, 5]
```

```python
my_list
```

```python
set_values = set(my_list)
```

```python
print(set_values)
```

```python
#convert the set to the list 
```

```python
unique_list = list(set_values)
```

```python
print(unique_list)
```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
```

```python
set1.union(set2)
```

```python
set1.intersection(set2)
```

```python
sub = {1, 3}
sub.issubset(set1)
```

```python
set1.issuperset(sub)
```

```python
#union
print(set1 | set2 )
```

```python
#intersection
print(set1 & set2 )
```
