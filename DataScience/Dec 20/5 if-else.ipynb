If the value given by a user is more then 10 then it should print value is more then 10 other wise print value is less than 10 


```python
x = int(input ("enter the value"))
if x > 10:
    print("The value is more then 10")
else:
    print("the value is less then 10 ")
    
```

```python
# check if the number is positive or negative
num = int(input("enter the value"))
if num > 0:
    print("It is a positive value")
elif num == 0:
    print("the value of num is zero")
else:
    print("it is a negative value")
```
