```python
import json


def clean_data(data):
    """
    This function cleans the input data by:
    1. Removing users with empty names
    2. Removing duplicate friends
    3. Removing inactive users
    4. Removing duplicate pages
    """

    # -----------------------------
    # 1️⃣ Remove users with missing / empty names
    # -----------------------------

    clean_users = []

    for user in data["users"]:
        name = user["name"]

        # remove spaces from start and end
        cleaned_name = name.strip()

        # keep user only if name is not empty
        if cleaned_name != "":
            clean_users.append(user)

    data["users"] = clean_users


    # -----------------------------
    # 2️⃣ Remove duplicate friends for each user
    # -----------------------------

    for user in data["users"]:
        friends_list = user["friends"]

        # remove duplicates using set
        unique_friends = set(friends_list)

        # convert back to list
        user["friends"] = list(unique_friends)


    # -----------------------------
    # 3️⃣ Remove inactive users
    # (users who have no friends AND no liked pages)
    # -----------------------------

    active_users = []

    for user in data["users"]:
        friends = user["friends"]
        liked_pages = user["liked_pages"]

        if friends or liked_pages:
            active_users.append(user)

    data["users"] = active_users


    # -----------------------------
    # 4️⃣ Remove duplicate pages
    # -----------------------------

    unique_pages = {}

    for page in data["pages"]:
        page_id = page["id"]

        # page_id becomes the key (ensures uniqueness)
        unique_pages[page_id] = page

    # convert dictionary values back to list
    data["pages"] = list(unique_pages.values())


    return data


# -----------------------------
# Load → Clean → Save data
# -----------------------------

# Load original JSON file
with open("codebook_data.json", "r") as file:
    data = json.load(file)
    print(data)

# Clean the data
cleaned_data = clean_data(data)

# Save cleaned data to new file
with open("cleaned_codebook_data.json", "w") as file:
    json.dump(cleaned_data, file, indent=4)

print("Data cleaned successfully!")
```
