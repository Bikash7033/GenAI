Dictionary in python is an unordered , mutable, and key value pair collection . It is allows efficient data retrieval and modification. Dictionary in python are ordered as of pyton 3.7


```python
#empty dictionary
empty_dic = {}
```

```python
# dictionary with key value pairs
student = {
    "name" : "Alice",
    "age" : 25,
    "grade": "A"
}
```

```python
print(student)
print(type(student))
```

```python
print(student["name"])
```

```python
print(student.get("gender", "Not found"))
```

```python
print(student.get("age", "Not found"))
```

```python
student = {
    "name": "Rahul",
    "skills": ["python", "ds", "Ml", "AI"]
}
print(student["skills"])
skills = student["skills"]
i = 0
while i < len(skills):
     print(skills[i])
     i = i + 1
```

```python
student = {
    "name": "Rahul",
    "skills": ["python", "ds", "Ml", "AI"]
}
print(student["skills"])
for i in student["skills"]:
  print(i)
```

```python
#common dictionary methods
#keys returns all keys in the dictionary
dict_key = student.keys()
list_keys = list(dict_key)
print(type(list_keys))
print(list_keys)
```

```python
# get all the values 
print(student.values())
```

```python
#items : returns key -value pair as tuple
# student.items() return 
# [ ("name", "Rahul"), ("skills", ["python", "ds", "ml"]
for key, value in student.items():
    print(key, value)
```
