JSON: java script object notation 
It is a lightweight data format used for data exchange between server and applications.
It is widely used in APIs , web applicatipons, and configurations

Python provides with the json module to work with json data. you can import the json module like :
import json 

Example of json:
 {
    "name": "Akshat"
    "age" : 50
    " skills" : ["python", "datascience"]
}




```python
import json
data = {
    "name": "alice",
    "age": 25,
    "city": "NYC"
}  #python dictionary
print(type(data))
# json.dumps()  ->>>>>>  convert python ---> json format 
json_string = json.dumps(data)  #convert the python dictionary to json and save in json_string
print(json_string)  #string string printed 
print(type(json_string))
#here python shows class as str because internally python stores json as a plain text 
```

#json.loads()   ->>>>> convert json ----> python format 
json.load()   ->>>>>> read json from a file to convert to python 
# create the data.json file 
{
   "name": "alice",
   "age" : 25,
   "skills": ["python", "aws"] 
}


```python
import json

file = open("data.json", "r") 
data = json.load(file)  #since here we are reading the data from file 
print(data)
print(type(data))
file.close()
```

```python
import json 
json_string = '{"name": "Alice", "age": 25, "skills": ["python", "aws"]}'
data = json.loads(json_string)
print(data)
print(type(data))
```
