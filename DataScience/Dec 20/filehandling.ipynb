File modes :
file = open("filename", mode)

"r" : Opens the file for reading and raise an error of the file does not exist
"w" : open the file for writing , create new file if not found and overwrite the existing file
"a" : Append : open the file for writing , creates a new file if not found  and appends content insted of overwriting
"x" : creates a new file , but fails if the file already exist 


```python
file = open("test.txt", "r")
content = file.read()
line1 = file.readline()
print(line1)
file.close()
```

```python
file = open("test.txt", "w")
file.write("hello akshat\n")
file.write("file handling\n")
file.close()
```

```python
file = open("test1.txt", "a")
file.write("New line added")
file.close()
```

```python
# using with open() ensures that the file is automatically closed after execution
with open("test1.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# before opening the file i want to check if the file exist or not 

import os
if os.path.exists("test1.txt"):
    print("file exists")
    with open("test1.txt", "r") as file:
     content = file.read()
     print(content)
else:
    print("file not found")
```

```python
# delete the file
import os

if os.path.exists("test1.txt"):
    os.remove("test1.txt")
    print("File deleted")
else:
    print("file does not exist")
```
