```python
candidates = ["Alice", "Bob", "Charlie"]
votes = [0, 0, 0]

while True:
    age = int(input("\nEnter your age (0 to stop voting): "))

    if age == 0:
        break

    if age < 18:
        print("Not eligible to vote")
        continue

    print("\nCandidates:")
    for i in range(len(candidates)):
        print(i + 1, "-", candidates[i])

    choice = int(input("Enter your vote (1-3): "))

    if 1 <= choice <= 3:
        votes[choice - 1] += 1
        print("Vote recorded")
    else:
        print("Invalid vote")
```

```python
print(votes)
max_votes = votes[0]  # assume the first candidate has max votes
winner = candidates[0] 

for i in range(len(votes)):
    if votes[i] > max_votes:
        max_votes = votes[i]
        winner = candidates[i]

print("\nFinal Results:")
for i in range(len(candidates)):
    print(candidates[i], ":", votes[i])

print("\nWinner is:", winner)
```
