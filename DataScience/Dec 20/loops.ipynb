```python
# print each character in a string
text = "helloworld"
for x in text:
    print(x)
```

```python
# count the characters in the string
text = "helloworld"
count = 0
for ch in text:
    count = count + 1 
print(count)
```

```python
# count the vowels in the string
text = "education"
vowels = "aeiou"
count = 0
for ch in text:
    print(ch)   # print the text as characters
    if ch in vowels:
        count = count + 1
print(f"vowels count {count}") 
```
