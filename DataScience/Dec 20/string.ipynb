```python
# count how many times "a" appears
sentence = "hello akshat we are good and we eat apple"
print(sentence.count("a"))  # print 4
print(f"count of a is {sentence.count("a")}")   # count of a is 6

# if you put f literal in the print statement then 
# you can just pass a variable in "" only 
# for example: print(f"the value of the variable is {x}")
# w/o f : print("the value of the variable is", x)

print("count of a is ", sentence.count("a"))  #count of a is 6
```

```python
# check if the string starts with word "hello"  and end with world
b = "hello akshat"
print(f"starts with hello: {b.startswith("hello")}")
print(f"starts with world: {b.endswith("world")}")
```

```python
# how to find the first and last word in the sentence
sentence = "python is a very easy langauge to learn"
words = sentence.split()
print(words)
print(words[0])
print(words[1])
print(words[2])
print(words[-1])
```
