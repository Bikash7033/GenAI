Object oriented programming (OOP) is programming paradigm  that organise the code into objects that contain both data 
(attributes) and behaviour (method) 

Class : A blueprint for creating objects 

A class is like the blueprint of the house 
the blueprint will tell you:
   - how many rooms
   - where the kitchen is
   - what is the size of house is
But the blueprint is not the house. it just defines how a house will look like.....

```python
class car:
    def __init__(self, brand, model):
        self.brand = brand   #attribute
        self.model = model   # attribute
    def display_info(self):  #method 
        return f"{self.brand} {self.model}"

#create an object
car1 = car("Tesla", "xyz")
print(car1.display_info())   
```

```python
class car:   #python create a blueprint
#ok !! i am creating a new class called car
#nothing runs yet. No object was created. Just a template is registered in memory
    def __init__(self, brand, model):  
#whenever someone creates a car object , this function must run automatically
#self -> will point to the new object
# brand , model -> values that must be provided
        self.brand = brand   #attribute    # car1.brand = “tesla”
        self.model = model   # attribute.  # car1.model = “xyz”
# python notes:
   # when __init___ runs :
      # store brand inside the object
      # store model inside the object


    def display_info(self):  #method 
#this class has a method called display_info. It will return brand + model 
        return f"{self.brand} {self.model}"

#create an object
car1 = car("Tesla", "xyz")
## what python actually do internally:
# python sees car(…)
#it will create an empty object of type car
#automatically calls :    __init__ (self = car1 , brand = “tesla”, “model”=xyz)

print(car1.display_info())  


# object : is a real thing created from the class that stores data 
# and can perform actions 

 
```

```python
# inheritance: It allows a class(child) to inherit attributes and method from another class(parent)
class Animal:
    def speak(self):
        return "Animal makes sound"
class Dog(Animal):
    def speak(self):
        return "Bark"

tommy = Dog()
print(tommy.speak())
```
