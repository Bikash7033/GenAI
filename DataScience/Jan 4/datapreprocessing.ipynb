```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Amit", "Akshat", None],
    "Age": [25,None, 35],
    "Marks": [80,90,None]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.isnull()
```

```python
#how many sum of all the Null values are there in each column
df.isnull().sum()
```

```python
# check if column has atlease one missing value 
df.isnull().any()
```

```python
#check if column is completely none (missing)
df.isnull().all()
```

```python
#find the total missing values in whole Data frame
df.isnull().sum().sum()
```

```python
# Drop missing  data ( We will use dropna())
df.dropna()  # it will drop all the rows which has Null entries 
```

```python
#Fillna : It is used to fill unknown values 
df.fillna(0)
```

```python
# If we want to drop a row if specific column has none
df.dropna(subset=["Age"])
```
