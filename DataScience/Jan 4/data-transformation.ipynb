```python
##DATA TRANSFORMATION : Once your data is clean, the next step is to reshape , reformat, and reorder it as  needed for analysis. 

###Pandas gives you plenty of flexible tools to do this 


import pandas as pd
data = {
    "Rollno": [1,2,3,4,5],
    "Name": ["Amit", "Ravi", "Neha", "Priya","Rohan"],
    "School": ["StJohn", "StJohn", "Xavier", "Xavier", "StJohn"],
    "Marks": [80,90,70,100,95],
    "email": ["amit@gmail.com","ravi@gmail.com","neha@gmail.com", "priya@gmail.com","rohan@gmail.com"]
}
df = pd.DataFrame(data)
print(df)
```

```python

df.sort_values("Marks")   #ascending order
```

```python

df.sort_values("Marks", ascending=False)
```

```python
df.sort_values(["Marks", "Name"])
```

```python
#rename a column
df.rename(columns={"email": "parent-email"}, inplace=True)
# df.rename(columns={oldnameofcolumn:newnameofcolumn}, inplace=True)
# inplace: modify the original object itself dont create a new one 
```
