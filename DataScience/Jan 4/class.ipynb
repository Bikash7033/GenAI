```python
import pandas as pd
data = {
    "Rollno": [1,2,3,4,5],
    "Name": ["Amit", "Ravi", "Neha", "Priya","Rohan"],
    "School": ["StJohn", "StJohn", "Xavier", "Xavier", "StJohn"],
    "Marks": [80,90,70,100,95],
    "email": ["amit@gmail.com","ravi@gmail.com","neha@gmail.com", "priya@gmail.com","rohan@gmail.com"]
}
df = pd.DataFrame(data)
print(df)
```

```python
#name starts with letter R
df.loc[df["Name"].str.startswith("R")]
```

```python
# studetns who have scored more then 85
df.loc[df["Marks"]>80]
```

```python
df.loc[(df["School"] == "StJohn") & (df["Marks"] > 85)]
```

```python
df.query("School == 'StJohn' and Marks > 85")
```

```python
df.query("School == 'StJohn' or Marks > 85")
```
