```python
# create an empty list and add elements to the list 
my_list = []
my_list.append(10)
print(my_list)
my_list.append(20)
print(my_list)
my_list.append(23)
print(my_list)
```

```python
# add multiple elements at once
num1 = [1, 2]
num1.extend([3, 4, 5])
print(num1)
```

```python
#insert element at specific position
num = [1,2]
num.insert(1,"python")
print(num)
```

```python
# remove the first occurence of 3 in the list 
num = [1, 2, 3, 4, 3]
num.remove(3)
print(num)
```

```python
# delete last element
num = [1, 2, 3, 4]
print(num)
num.pop()
print(num)
```

```python
#clear all the elements
num = [1, 2, 3, 4]
print(num)
num.clear()
print(num)
```
