```python
# I WANT TO FIND IF THE NUMBER IS EVEN OR ODD VIA FUNCTION
def evenorodd(a):
   if a%2 == 0:
       return  "Even"
   else:
        return "Odd"

x = evenorodd(int(input("enter a value")))
print(f"We have the value {x}")
```

```python
# write a function to check if the email are valid or not 
def emailvalid(x):
    if "@" in x and "." in x:
        return "valid email"
    else:
        return "invalid email"

print(emailvalid("akshatawsac@gmailcom"))
```
