```python
#buying behaviour of the users as per there age and salary
from sklearn import tree
salary_age = [[25, 30000], [30, 40000], [35, 60000], [40,80000], [22,20000], [45,90000]]
#0 -> no (did not buy)
#1 -> yes ( purchased)
purchased = [0, 0, 1, 1, 0, 1]
model = tree.DecisionTreeClassifier()
model.fit(salary_age, purchased)
print("model is trained successfully ")
prediction = model.predict([[ 60, 95000]])
print(prediction)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
#predict the salary based on years of experience (linear regression)
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([30000, 35000, 40000, 45000, 50000])  #salary
model = LinearRegression()
model.fit(x,y)
prediction = model.predict([[6]])
print(prediction)
```

```python
#logistic regression -> predict pass or fail based on hours studied
from sklearn.linear_model import LogisticRegression
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 0, 1, 1]) # 0 -> fail and 1 -> pass
model = LogisticRegression()
model.fit(x,y)
print(model.predict([[8]]))
```

```python
#predict pass or fail using marks + attendance
from sklearn.ensemble import RandomForestClassifier
x = np.array([[35,60], [40,65], [60,80], [80, 90]])
y = np.array([0, 0, 1, 1])
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x,y)
print(model.predict([[55, 75]]))
```
