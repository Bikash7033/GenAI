## Structured Prompt Formula

Role + Task + Context + Format + Constraints + Output

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

prompt = """
Role: Act as  senir QA Automation Trainer

Task: Explain API Testing.

Context: Students are manual testers and want to learn API testing. 
They have no prior experience in API testing. They are eager to learn and want to understand the basics of API testing.

Format: 

1. Simple definition.
2. Real-world example.
3. Why it is important.
4. Practice task.

Constraints:
- Use simple language. Avoid technical and heavy jargon.

"""

response = client.responses.create(
    model="gpt-5.4-nano-2026-03-17",
    input=prompt
)

print(response.output_text)
```

## Simple Chatbot

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

print("AI Chatbot started. Type exit to stop the chatbot.")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot stopped.")
        break

    response = client.responses.create(
        model="gpt-4o-mini",
        instructions="You are a helpful and friendly chatbot. Answer the user's questions and engage in a conversation.",
        input=user_input
    )

    print("Chatbot:", response.output_text)
```

## Decomposition Prompting

Decomposition means breaking a big topic into smaller parts.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

topic = "Python programming"

prompt = """
Help in learning {topic} by rreaking it into smaller subtopics and providing a step-by-step learning path for each subtopic.

use this structure:

1. What is programming?
2. What is Python Programming?
3. Why Python is popular?
4. Basic Syntax and Data Types
5. Control Structures (if, for, while)
6. Functions and Modules
7. Object-Oriented Programming
8. Libraries and Frameworks

User is a beginner and wants to learn Python programming from scratch. Provide simple explanations and examples for each subtopic.
"""

response = client.responses.create(
    model="gpt-4o-mini",
    input=prompt
)


print(response.output_text)
```

## Self Consistency / Multiple Outputs

instead os asking once asnwer, we will ask Ai to generate multiple optinos andcompare them

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

prompt = """
I want to learn Python.

Genrate 3 different learning plans:
1. Fast-Track plan
2. beginner-friendly plan
3. Job oriented plan

For each plan, include the following details:
- Duration: [Estimated time to complete the plan]
- Topics Covered: [List of topics included in the plan]
- Practice Method: [Recommended practice methods, such as projects, exercises, or quizzes]
- Best for whom: [Describe the ideal learner for this plan, such as absolute beginners, those with some programming experience, or those looking to get a job in Python development]

Then recommend the best plan for a working professional.

"""

response = client.responses.create(
    model="gpt-4o-mini",
    input=prompt
)

print(response.output_text)
```

## Tree of Thought prompting

ToT means Ai explores multiple paths before recommending the best one.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))


prompt = """
I have 10 years of QA/SDET experience and want to move into AI

compare 3 career paths for me:
1. AI Developer
2. AI Automation Engineer
3. AI Research Scientist

Fr each path, analyze:
-Required skilss
- Difficulty level
- Time to transition
- Salary potential as per market trends in India and globally
- Risk
- Best first step for transition

Then recommend the best path
"""

response = client.responses.create(
    model="gpt-4o-mini",
    input=prompt
)

print(response.output_text)
```
