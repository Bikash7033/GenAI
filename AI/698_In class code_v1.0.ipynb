```python
hamlet=nltk.corpus.gutenberg.words('shakespeare-hamlet.txt')
```

```python
len(hamlet)
```

```python
hamlet_pos = []
```

```python
for word in hamlet[:2000]:
    word_pos=nltk.pos_tag([word])
    hamlet_pos.append(word_pos)
```

```python
hamlet_pos
```

```python
hamlet_nnp = []
for each_pos in hamlet_pos:
    if each_pos[0][1] =='NNP': #proper noun
        hamlet_nnp.append(each_pos[0][0])
```

### CHUNKING

Grouping words into meaningful phrases

instead of seeing individaula words, the system sees phrases

NP  Noun phrase
VP  Verb Phrase
PP prepositional phrase

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python
from nltk import ne_chunk
```

```python
NE_sent = "The US President stays in the White House"
```

```python
NE_tokens = word_tokenize(NE_sent)
```

```python
NE_tags = nltk.pos_tag(NE_tokens)
```

```python
NE_NER = ne_chunk(NE_tags)
```

```python
print(NE_NER)
```

```python
NE_sent2 = "The state of New York touches the Atlantic Ocean"
```

```python
print(ne_chunk(nltk.pos_tag(word_tokenize(NE_sent2)))) #ner
```

```python
NE_sent3 = "Apple is a fruit and Apple is a Company's name"
```

```python
print(ne_chunk(nltk.pos_tag(word_tokenize(NE_sent3))))
```
