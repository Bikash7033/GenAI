```python
# POS tagging + Lemmatization

# map the tags to tags which lemmatizer understands
def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
get_wordnet_pos('NN')
```

```python
words
```

```python
lemmatizer = WordNetLemmatizer()

for word, tag in nltk.pos_tag(words):
        lemma = lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        print(f"{word}-{lemma}")
```

### TF-IDF

Idea:

* Common words like "the", "is" they should have less importance
* Rare words should have more importance

TF-IDF = TF * IDF

TF= term frequency (word)

IDF= inverse term frequency

TF(t, d) = **Number of times the term t appears in a document d/ Total number of words in document d**

Document1: "I love NLP and I love machine learning"

TF(love, d=1) = 2/8 = 0.25

TF(NLP, d=1) = 1/8 = 0.125

So more frequent words -> higher TF

classification
1000 docs/ articles

TF(cricket, d=1) =.38 #document no. 1...most likey is about cricket

TF(hockey, d=1) = 0

REview example)

TF(good, R1)= .34

TF(not, R1) = .28

TF(movies, R1) = .5 #movie is not an impactful word for our use case

**IDF**

Inverse Document Frequency

IDF measures how rare a word is across all documents

IDF(t) = log (N/df(t))

N= total number of documents
df(t) = number of documents containing the term t

D1: I love NLP

D2: I love AI

D3: AI is the future

Vocabulary
[I, love, NLP, AI, is, the, future]

IDF(love) = log(3/2) = 0.176

IDF(NLP) = log(3/1) = 0.47

Rare words -> higher IDF
Coomon words -> lower IDF

filler/ stop words/ movie -> high IDF or low IDF? low idf

**FINAL TF-IDF Score**

TF*IDF

TF_IDF(NLP, D1) = TF(NLP, D1) * IDF(NLP) = 0.125*0.47
TF_IDF(NLP, D2) = TF(NLP, D2) * IDF(NLP) = 0*0.47

frequent in the document
rare across the dataset
-> high importance score
