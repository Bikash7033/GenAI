Dataset used: **SMS Spam Collection Dataset (UCI Machine Learning Repository)**.

We will implement the pipeline twice:

1. **Bag of Words + Naive Bayes**
2. **TF-IDF + Naive Bayes**

## Step 1 — Import Libraries

We use common libraries from the Python NLP ecosystem.

- pandas → data handling
- scikit‑learn → ML models & vectorizers
- matplotlib / seaborn → visualization


```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB #one of the ML models
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2 — Load Real Dataset

We use the **SMS Spam Collection Dataset**.

Source:
https://archive.ics.uci.edu/ml/datasets/sms+spam+collection

Each message is labeled as:

- ham → normal message
- spam → unwanted promotional message


```python
url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"

data = pd.read_table(url, header=None, names=['label','message']) #pandas function read_table

data.head()
```

We convert labels into numeric values for the ML model.

- ham → 0
- spam → 1


```python
data['label_num'] = data.label.map({'ham':0,'spam':1}) #

data.head()
```

## Step 4 — Train/Test Split

We split the dataset so that the model is evaluated on **unseen messages**.

Training dataset: we are going to train ML model
Test datasets: check the performce of the model (unseen data)

```python
X_train, X_test, y_train, y_test = train_test_split( #randomly splits
    data['message'], #X(features -> message)
    data['label_num'], #Y(labels  -> spam/ham)
    test_size=0.2, #20% of data is test data
)
```

# PART 1 — Bag of Words

Bag of Words converts text into a matrix of **word frequencies**.

Steps:

1. Build vocabulary (unique words in training dataset)
2. Count word occurrences
3. Create document‑term matrix

```python
bow_vectorizer = CountVectorizer(stop_words='english')

X_train_bow = bow_vectorizer.fit_transform(X_train) #Features of training data; 4457 SMS for training
X_test_bow = bow_vectorizer.transform(X_test)

# Why are we applying fit_transform method on training data and transform method on test data

X_train_bow.shape # 4457 SMS messages (size of training data); for each SMS we are creatig a numerical vector of size 7460(vocabulary of the model)
```

## Train Naive Bayes Model

```python
model_bow = MultinomialNB() #initiakizing NB class
model_bow.fit(X_train_bow, y_train) #trains the model (features, labels[0 for ham and 1 for spam])

pred_bow = model_bow.predict(X_test_bow)

print("Accuracy:", accuracy_score(y_test, pred_bow)) #y_test-> actual labels , pred_bow-> predicted labels
```

## Evaluation

```python
print(classification_report(y_test, pred_bow))
```

# PART 2 — TF‑IDF

TF‑IDF improves Bag of Words by giving **less importance to common words**.

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
X_train_tfidf.shape
```

## Train Naive Bayes Model

### Naive Bayes Model -> classification problem

DURING THE TRAINING

Training data: 1000 emails

Email: Free stuff.

P(spam) = P(spam/ free) * P(spam/ stuff) = 0.8

0.8 > Threshold(0.5) -> Tag it as spam
prob < threshold -> tag it as ham

P(A/B) = P(A) * P(B/A) / P(B)

**P(spam/free) = P(spam) * P(free/spam) / P(free)**

1000 emails  -> 6000 unique words (Vocabulary of the model)

1000 emails -> 600 ham mails and 400 spam mails

out of 600 ham mails, lets say...100 mails have word "free" in it

out of 400 spam mails, lets say ...200 mails have word "free" in it

P(spam) = 400/1000 = 0.4 (40% percent)

P(free) = 300/1000 = 0.3

P(free/spam) = What is the probability that a spam has word free given it is a spam mail. = 200/400 = 0.5

P(spam/free) = 0.4 * 0.5/0.3 = 2/3 = 0.66

You will repeat same exercise for all the words in vocabulary of your model

P(spam/w1), P(spam/w2)..............P(spam/w6000)

Training of the model is....during training it learns these conditional probabilities....

PREDICTION

mail: free goodies, hurry up!

P(spam) = p(spam/free) * p(spam/goodies) * P(spam/hurry) *P(spam/up)

P(spam) > Threshold -> SPAM mail

P(spam) < Threshold -> HAM mail

```python
model_tfidf = MultinomialNB() #if i want to use threshold different from 0.5...i have to do it externally
model_tfidf.fit(X_train_tfidf, y_train)

pred_tfidf = model_tfidf.predict(X_test_tfidf) #threshold of 0.5

print("Accuracy:", accuracy_score(y_test, pred_tfidf))
```

```python
#TO CHANGE THE THRESHOLD...do it externally
probs = model_tfidf.predict_proba(X_test_tfidf)

threshold = 0.7
y_pred_70 = (probs[:,1] > threshold).astype(int)
```

## Evaluation

```python
print(classification_report(y_test, pred_tfidf))
```
